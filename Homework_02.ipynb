```python
# Implement the cost function here
theta = np.array([0.0, 0.0, 0.0])

X_data = df[['x1', 'x2']].values
X = np.hstack((np.ones((len(df), 1)), X_data))
y = df['label'].values
m = len(y)

def cal_cost(theta, X, y):
  htheta = sigmoid(np.matmul(X, theta))
  term1 = np.dot(y, np.log(htheta))
  term2 = np.dot((1-y), np.log(1-htheta+1e-15))
  J = - 1 / m * (term1 + term2)
  return J

def cal_grad(theta, X, y):
  htheta = sigmoid(np.matmul(X, theta))
  term1 = np.matmul(X.T, (htheta - y))
  grad = 1 / m * term1
  return grad

print('Cost at initial theta :', cal_cost(theta, X, y))
print('Gradient at inital theta :', cal_grad(theta, X, y))
```

```python
from matplotlib.collections import EllipseCollection
# Gradient Descent Implementation Here
learning_rate = 1e-3
epochs = 80000

# Re-initialize theta to zeros so we start training from scratch
theta = np.array([0.0, 0.0, 0.0])

for i in range(epochs):
  gradient = cal_grad(theta, X, y)
  theta -= learning_rate * gradient

  if i < 5:
    print(f"------Epoch {i}------")
    print(f"Theta: {theta}")
    cost = cal_cost(theta, X, y)
    print(f"Cost: {cost}")

final_cost = cal_cost(theta, X, y)
print(f"\nBest Theta: {theta}")
print(f"Cost: {final_cost}")
```


### 1.7 - Regularization (L1 and L2)

Practice adding L1 and L2 regularizers to your logistic regression implementation. **(15 points)**

**Tasks:**
1. Add a regularization coefficient `lambda_` to your training code.
2. Implement **L2** regularization by shrinking only the non-bias parameters (`theta[1:]`). (**Do not** regularize `theta[0]`.)
3. Implement **L1** regularization by applying a sub-gradient to only the non-bias parameters (`theta[1:]`).
4. Train with a few values of `lambda_` (e.g., 0.1, 1) and compare the learned theta and decision boundary.

**Hints:**  
1. L2 (weight decay) update: `theta[1:] = theta[1:] - alpha * (grad[1:] + (lambda_/m) * theta[1:])`.
2. L1 update (sub-gradient): `theta[1:] = theta[1:] - alpha * (grad[1:] + (lambda_/m) * np.sign(theta[1:]))`.
3. If you also compute a regularized cost for monitoring, regularize only `theta[1:]` (not `theta[0]`).


```python
# Practice: regularize only theta[1:] with L2 and retrain here.
lambda_ = 1.0
alpha = 0.001
epochs = 80000
theta_l2 = np.array([0.0,0.0,0.0])

for i in range(epochs):
    grad = cal_grad(theta_l2, X, y)
    theta_l2[0] -= alpha * grad[0]
    theta_l2[1:] -= alpha * (grad[1:] + (lambda_ / m) * theta_l2[1:])

print(f"L2 Regularization (lambda={lambda_})")
print(f"Theta: {theta_l2}")
print(f"Unregularized Cost: {cal_cost(theta_l2, X, y)}")

plt.plot(x_boundary, y_boundary, label='L2 Boundary', color='orange')
colors = np.where(df['label']==0, 'red', 'blue')
plt.scatter(df['x1'], df['x2'], c=colors)
plt.show()
```

```python
# Practice: regularize only theta[1:] with L1 and retrain here.
lambda_ = 1.0
alpha = 0.001
epochs = 80000
theta_l1 = np.array([0.0,0.0,0.0])


for i in range(epochs):
    grad = cal_grad(theta_l1, X, y)
    theta_l1[0] -= alpha * grad[0]
    theta_l1[1:] -= alpha * (grad[1:] + (lambda_ / m) * np.sign(theta_l1[1:]))

print(f"L1 Regularization (lambda={lambda_})")
print(f"Theta: {theta_l1}")
print(f"Unregularized Cost: {cal_cost(theta_l1, X, y)}")

# Plot Decision Boundary
colors = np.where(df['label']==0, 'red', 'blue')
plt.scatter(df['x1'], df['x2'], c=colors)

plt.plot(x_boundary, y_boundary, label='L1 Boundary', color='orange')
plt.show()
```
